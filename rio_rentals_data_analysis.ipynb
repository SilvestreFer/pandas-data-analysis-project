### Conditional Filtering

In this section, we’ll apply filters to identify:
1. Apartments with one bedroom and rent lower than 1200.  

```python
# Returns True for apartments with one bedroom, otherwise False.
apartments_df['Quartos'] == 1
```

```python
# Filtering apartments with one bedroom.
one_bedroom_filter = apartments_df['Quartos'] == 1
apartments_df[one_bedroom_filter]
```

```python
# Filtering apartments with rent lower than 1200.
low_rent_filter = apartments_df['Valor'] < 1200
apartments_df[low_rent_filter]
```

```python
# Combining both conditions.
one_bedroom_low_rent_apartments = apartments_df[one_bedroom_filter & low_rent_filter]
# Displaying the resulting DataFrame.
one_bedroom_low_rent_apartments.head()
```

2. Apartments with at least two bedrooms, rent lower than 3000, and area greater than 70.

```python
# Filtering apartments with at least two bedrooms, rent lower than 3000, and area greater than 70.
large_affordable_apartments_filter = (
    (apartments_df['Quartos'] >= 2) &
    (apartments_df['Valor'] < 3000) &
    (apartments_df['Area'] > 70)
)

# Applying the filter
large_affordable_apartments = apartments_df[large_affordable_apartments_filter]

# Displaying the resulting DataFrame
large_affordable_apartments.head()
```

3. Apartments with area greater than 80 m² **or** rent lower than 4000.

```python
# Boolean filter: apartments with area > 80 or rent < 4000
large_or_low_rent_filter = (apartments_df['Area'] > 80) | (apartments_df['Valor'] < 4000)

# Applying the filter
large_or_low_rent_apartments = apartments_df[large_or_low_rent_filter]

# Displaying the filtered DataFrame
large_or_low_rent_apartments
```

## **MANIPULATING DATA**

### **Reloading the Data**

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
rental_df = pd.read_csv(url, sep=';')
rental_df.head()
```

### **Creating Numeric Columns**

```python
rental_df['Valor_por_mes'] = rental_df['Valor'] + rental_df['Condominio']
rental_df.head()
```

```python
rental_df['Valor_por_ano'] = rental_df['Valor_por_mes'] * 12 + rental_df['IPTU']
rental_df.head()
```

### **Creating Categorical Columns**

```python
rental_df['Descricao'] = rental_df['Tipo'] + ' em ' + rental_df['Bairro']
rental_df.head()
```

```python
rental_df['Descricao'] = rental_df['Tipo'] + ' em ' + rental_df['Bairro'] + ' com ' + \
                     str(rental_df['Quartos']) + ' quarto(s) ' + 'e ' + \
                     str(rental_df['Vagas']) + ' vaga(s) de garagem.'
rental_df.head()
```

```python
rental_df['Descricao'] = rental_df['Tipo'] + ' em ' + rental_df['Bairro'] + ' com ' + \
                                        rental_df['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + rental_df['Vagas'].astype(str) + ' vaga(s) de garagem.'
rental_df.head()
```

```python
rental_df['Possui_suite'] = rental_df['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
rental_df.head()
```
